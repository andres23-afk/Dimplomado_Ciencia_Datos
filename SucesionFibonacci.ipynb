La sucesión de Fibonacci es una serie de números descubiertos por Leonardo Fibonacci. Consiste en la suma de los dos números anteriores de la sucesión. Los números se utilizan para crear patrones y fórmulas. Esencialmente, se utilizan para construir estructuras más complejas.

Fibonacci fue un matemático y filósofo italiano de treinta y dos años que vivió entre 1175 y 1250. Aunque vivió hace muchos siglos, su trabajo todavía se conoce a menudo. Introdujo la idea de describir la naturaleza usando patrones numéricos. Además, introdujo la idea de describir y clasificar la naturaleza mediante fórmulas numéricas. Llamó a estas ideas 'logística' y 'numerología'. Hoy llamamos matemáticas a este tipo de trabajo.

La secuencia que conocemos como la secuencia de Fibonacci recibió su nombre de Leonard Fibonacci, según Fibonacci, cada número de la secuencia se genera sumando dos números de la secuencia anterior. Además, cada número es uno mayor que su predecesor. Por ejemplo, si el primer número de una serie es 1, el siguiente número será 2 porque 1 más 1 es 2. Después de eso, el siguiente número será 3 porque 1 más 2 es 3. Y así sucesivamente para cada número subsiguiente. en la secuencia.

Fibonacci usó estos números para crear estructuras elaboradas y formar órganos humanos como corazones y oídos. También los usó para calcular patrones de crecimiento en plantas y el desarrollo del sistema muscular de Amaranthaceae en peces. Además, estos números se pueden usar para generar números aleatorios, lo que los hace útiles para juegos como los dados o los juegos de casino.

En general tenemos que: \
$F_{n} = \tfrac{(1 + \sqrt{5})^{n} - (1 - \sqrt{5})^{n}}{2^{n} \sqrt{5}}$

El siguiente código nos da una muestra de la sucesión de Fibonacci:

```python
from math import sqrt
def F(n):
    return ((1+sqrt(5))**n-(1-sqrt(5))**n)/(2**n*sqrt(5))

def Fibonacci(startNumber, endNumber):
    n = 0
    cur = F(n)
    while cur <= endNumber:
        if startNumber <= cur:
            print(cur)
        n += 1
        cur = F(n)
Fibonacci(1,100)
```
